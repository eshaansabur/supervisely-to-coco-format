```python
import json
import os

# Specify the directory containing the JSON files
directory = "C:\Projects\Pigeon to Coco"

# Create an empty list to store the JSON data
data_list = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    # Check if the file is a JSON file
    if filename.endswith(".json"):
        # Construct the full file path
        print(filename)
        # Extract the filename without the extension
        extracted_filename = filename.split(".json")[0]
        print(extracted_filename)
        filepath = os.path.join(directory, filename)

        # Open the JSON file
        with open(filepath, "r") as f:
            # Load the JSON data
            data = json.load(f)

            # Append the JSON data to the list
            data_list.append({"filename": extracted_filename, "data": data})

print(data_list)
```

```python
import datetime
import numpy as np
for data in data_list:
    demo_data= data['data']
    #print(demo_data)
    coco_description = demo_data['description']
    #print(coco_description)
    first_data = demo_data['objects'][0]
    contributor = first_data['labelerLogin']
    print(contributor)
    #print(first_data)
    coco_date = first_data.get('createdAt')
    #print(coco_date)
    # Extract date and time
    date_time = datetime.datetime.strptime(coco_date, "%Y-%m-%dT%H:%M:%S.%fZ")
    
    # Extract date
    year = date_time.year
    #print(year)
    date = date_time.date()
    
    # Format date
    formatted_date = date.strftime("%Y-%m-%d")
    #print(formatted_date)
    #info objects
    info_object= {
    
    }
    info_object.update({"description": coco_description})
    info_object.update({"year": year})
    info_object.update({"version": 1})
    info_object.update({"contributor":  contributor})
    info_object.update({"url": ""})
    info_object.update({"date_created": formatted_date})
    #print(info_object)
    #licenses
    licenses = []
    licenses.append({"id": 1, "name": "Attribution-NonCommercial", "url": "https://creativecommons.org/licenses/by-nc/2.0/"})
    #print(licenses)
    #images
    images = []
    # Extract the filename without the extension
    extracted_filename = data['filename']
    #print(extracted_filename)
    images_id= 1
    images.append({"id": images_id,"width": demo_data['size']['width'], "height": demo_data['size']['height'], "file_name": extracted_filename, "license": licenses[0]['id']})
    #print(images)
    annotation = []
    objects = demo_data['objects']
    #print(objects)
    categories_array = []
    for object in objects:
        #print(object['classTitle'])
        if object['classTitle'] not in categories_array:
            categories_array.append(object['classTitle'])
    
    #print(categories_array)
    categories_object_array = []

    for i, category in enumerate(categories_array):
        dict = {"id": i+1, "name": category}
        categories_object_array.append(dict)
    
    #print(categories_object_array)
    id =1
    for object in objects:
        #print(object['classTitle'])
        for i, category in enumerate(categories_array):
            if category == object['classTitle']:
                category_id = i+1
                #print(category_id)
        
        # Define the points
        points= np.array(object['points']['exterior'])
        #print(points)
        # Extract x1, y1, x2, y2
        x1 = points[0, 0]
        y1 = points[0, 1]
        x2 = points[1, 0]
        y2 = points[1, 1]
        x3= x2-x1
        y3= y2-y1
        bbox = [int(x1), int(y1), int(x3), int(y3)]
        #print(bbox)
        area = bbox[2] * bbox[3]
        #print(area)
        isCrowd =0
        annotation_appender= {"id": id, "image_id": images_id, "category_id": category_id, "bbox": bbox, "area": area, "isCrowd": isCrowd}
        #print(annotation_appender)
        annotation.append(annotation_appender)
        id=id+1
    #annotation
    #print(annotation)
    coco_format = {"info": info_object, "licenses": licenses, "images": images, "annotation": annotation, "categories": categories_object_array}
    #print(coco_format)
    # Combine the filename with the base filename to create the full filename
    formatted = "cocoformatted" + extracted_filename+".json"
    print(formatted)
    # Specify the directory name
    directory_name = "coco_format_data"
    
    # Create the directory if it doesn't exist
    if not os.path.exists(directory_name):
        os.makedirs(directory_name)
    # Construct the full file path
    filepath = os.path.join(directory_name, formatted)
    # Save the JSON data to the file
    with open(filepath, "w") as f:
        json.dump(coco_format, f, indent=4)
    images_id= images_id+1
    
    
```
